```python
file_path = r"C:\Users\rezgu\OneDrive\Bureau\Housing_pricing\data\raw\dataset_clean.csv"
df = pd.read_csv(file_path)
df.head(3)
```

```python
df['governorate'].unique()
```

```python
def process_governorate(row):
    if row['governorate'] == 'tunis':
        return 'tunis'
    if row['governorate'] == 'Ariana':
        return 'ariana'
    elif row['governorate'] == 'Ben Arous':
        return 'ben arous'
    elif row['governorate'] == 'La Manouba':
        return 'manouba'
```

```python
def isFromGrandTunis(governorate):
    return governorate.lower() in ["tunis", "ariana", "ben_arous", "ben arous", "manouba", "la manouba", "la_manouba"]
```

```python
df = df[df["governorate"].apply(isFromGrandTunis)]
df['governorate'].unique()
```

```python
df['governorate']=df.apply(process_governorate,axis=1)
df.head(4)
```
