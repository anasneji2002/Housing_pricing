# **Format TechnocasaDataset.csv price column**

### **What to Expect ?**

This notebook aims to do the following 
 - Change the formatting of prices for technocasa dataset so that it can be float parseable (now it has the format of XX XXX DT which isn't float parseable)

```python
import pandas as pd
```

```python
def formatPrice(price):
    return price.replace(" ", "").replace("DT", "")

FILE_PATH = 'C:\\Users\\mohamedanas.neji\\OneDrive - Medius\\Desktop\\Housing_pricing\\data\\raw\\TechnocasaDataset.csv'
```

```python
data = pd.read_csv(FILE_PATH)
data.head(3)
```

```python
data["Price"] = data["Price"].apply(formatPrice)
data.head(3)
```

```python
# Save changes
data.to_csv(FILE_PATH)
```
