# Mubaweb dataset update

This file contains the update of the columns order in mubaweb dataset to meet the template


```python
import pandas as pd
```

```python
def transform_csv(input_file, output_file):
    """
    Transform CSV file by renaming, reordering and selecting specific columns.
    
    Parameters:
    input_file (str): Path to input CSV file
    output_file (str): Path to output CSV file
    """
    # Read the CSV file
    df = pd.read_csv(input_file)
    
    # Create column mapping dictionary
    column_mapping = {
        'Surface': 'surface',
        'Rooms': 'rooms',
        'Bathrooms': 'bathrooms',
        'garden': 'jardin',
        'Price': 'price',
        'Location': 'city',
    }
    
    # Rename the columns
    df = df.rename(columns=column_mapping)
    
    # Select and reorder columns
    selected_columns = [
        'surface',
        'city',
        'rooms',
        'bathrooms',
        'parking',
        'pool',
        'vue_panoramique',
        'jardin',
        'climatisation',
        'chauffage_central',
        'ascenseur',
        'price'
    ]
    
    # Create new dataframe with selected columns
    df_transformed = df[selected_columns]
    
    # Save to new CSV file using comma as separator
    df_transformed.to_csv(output_file, sep=',', index=False)
    
    return df_transformed
```

```python
transform_csv("../data/raw/mubaweb.csv", "../data/raw/mubaweb_transformed.csv")
```
