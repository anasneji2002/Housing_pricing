# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# 1. Visualizing Price and Size Distribution (2D)
plt.figure(figsize=(10, 6))

# Plot distribution of Price
plt.subplot(1, 2, 1)
sns.histplot(df['price'], kde=True, color='blue')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Plot distribution of Size (Surface)
plt.subplot(1, 2, 2)
sns.histplot(df['size'], kde=True, color='green')
plt.title("Size Distribution")
plt.xlabel("Size (Surface)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# 2. Price vs Region Distribution
plt.figure(figsize=(10, 6))

# Boxplot for price distribution by region
sns.barplot(x='region', y='price', data=df, palette="Set2")
plt.title("Price Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Price vs surface Distribution
# Scatter plot for Price vs Surface (Size)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size', y='price', data=df, color='blue', s=100, alpha=0.7)
plt.title("Price vs Surface Distribution")
plt.xlabel("Surface (Size)")
plt.ylabel("Price")
plt.show()
```

```python
# Regression plot for Price vs Surface (Size)
plt.figure(figsize=(10, 6))
sns.regplot(x='size', y='price', data=df, scatter_kws={'s': 100, 'alpha': 0.7}, line_kws={'color': 'red'})
plt.title("Price vs Surface (Size) with Regression Line")
plt.xlabel("Surface (Size)")
plt.ylabel("Price")
plt.show()
```

```python
# 3. 3D Plot: Price vs Size vs Region
# Encode region as categorical numerical values for 3D visualization
df['region_code'] = df['region'].astype('category').cat.codes

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='size', y='price', z='region_code',
                    color='region',
                    labels={'size': 'Surface Size', 'price': 'Price', 'region_code': 'Region'},
                    title="3D Plot: Price vs Size vs Region")
fig.show()
```
