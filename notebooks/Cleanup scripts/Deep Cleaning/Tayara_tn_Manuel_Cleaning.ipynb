## Convert unit to TND and correct some mistakes

```python
# Filtering rows with price in the range of 100 to 999
price_filtered_df = clean_df[(clean_df['price'] >= 100) & (clean_df['price'] <= 999)]
print (len(price_filtered_df))
# Multiply the price by 100 for these rows
clean_df.loc[price_filtered_df.index, 'price'] *= 100
# Filtering rows with price over 5 Billion dinar
price_filtered_df2 = clean_df[clean_df['price'] >= 5000000]
print (len(price_filtered_df2))
# Divide the price by 100 for these rows
clean_df.loc[price_filtered_df2.index, 'price'] /= 100
```

```python
# Set the correct price
if 647 in clean_df.index:
    clean_df.at[647, 'price'] = 420000
if 648 in clean_df.index:
    clean_df.at[648, 'price'] = 315000
if 588 in clean_df.index:
    clean_df.at[588, 'price'] = 2300*136
```

```python
# Find rows where the price is not a multiple of 10^5
not_multiple_of_10_5_df = clean_df[(clean_df['price'] // 10**5 == 0) & (clean_df['price'] // 10**4 != 0)   ]

display(not_multiple_of_10_5_df)
```

```python
# Multiply the price by 10 the price except for indexes 207, 302, 692, and 694
indexes_to_exclude = [207, 302, 692, 694]
indexes_to_update = not_multiple_of_10_5_df.index.difference(indexes_to_exclude)
clean_df.loc[indexes_to_update, 'price'] *= 10
```

```python
# Find rows where the price is not a multiple of 10^4
not_multiple_of_10_4_df = clean_df[(clean_df['price'] // 10**4 == 0)  ]

display(not_multiple_of_10_4_df)
```

```python
# Update the price feature by multiplying the current value with the surface feature value for specified rows
for index in not_multiple_of_10_4_df.index:
    if index in df.index:
        clean_df.at[index, 'price'] *= float(clean_df.at[index, 'surface'])
```
