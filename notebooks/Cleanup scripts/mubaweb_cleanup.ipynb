# Mubaweb.csv cleanup

### this cleanup file contains:
 - formatting: 
   - price
   - rooms
   - bathrooms
   - surface
 - deducing the values of :
    parking, pool, vue_panoramique, jardin, climatisation, chauffage_central and ascenseur
    from the 'Amenities' column
 - removing unecessary columns      

```python
import pandas as pd
```

```python
file_path = r"C:\Users\rezgu\OneDrive\Bureau\Housing_pricing\data\raw\mubaweb.csv"
df = pd.read_csv(file_path)
```

```python
df.columns
```

```python
df['Price']=df['Price'].str.replace('TND','')
df['Price']=df['Price'].str.replace('\xa0','')
df['Price'].astype(float)
df['Price'].head()
```

```python
df['Surface']=df['Surface'].str.replace('m²','')
df['Surface'].astype(float)
df['Surface'].head()
```

```python
df = df[(df['Rooms'] != 'No number of rooms') & (df['Bathrooms'] != 'No number of bathrooms')]
```

```python
df['Rooms']=df['Rooms'].str.replace(' Chambres','')
df['Rooms']=df['Rooms'].str.replace(' Chambre','')
df['Rooms'].astype(float)
df['Rooms'].head()
```

```python
df['Bathrooms']=df['Bathrooms'].str.replace(' Salles de bains','')
df['Bathrooms']=df['Bathrooms'].str.replace(' Salle de bain','')
df['Bathrooms'].astype(float)
df['Bathrooms'].head()
```

```python
# List of amenities 
amenities_list = {
    'parking': 'Garage',
    'pool': 'Piscine',
    'vue_panoramique': 'Vue sur les montagnes',
    'vue_sur_mer': 'Vue sur mer',
    'jardin': 'Jardin',
    'climatisation': 'Climatisation',
    'chauffage_central': 'Chauffage central',
    'ascenseur': 'Ascenseur'
}

# Function to check if the amenity is in the 'amenities' list
def check_amenity(amenities, amenity):
    if pd.isna(amenities):  # Check if the value is NaN
        return False
    return amenity in amenities



```

```python

description_list={
    'parking': 'parking',
    'pool': 'piscine',
    'vue_panoramique': 'vue panoramique',
    'jardin': 'jardin',
    'climatisation': 'climatiseur',
    'chauffage_central': 'chauffage',
    'ascenseur': 'ascenseur'
}
def check_descriptions(description,amenity):
    if pd.isna(description):
        return False
    return amenity in description
```

```python

# Loop through the amenities and create columns
for column, amenity in amenities_list.items():
    df[column] = df['Amenities'].apply(lambda x: check_amenity(x, amenity)) 
   

for column, amenity in description_list.items():
    df[column] = ( df[column] | df['Description'].apply(lambda x: check_descriptions(x, amenity)))    
```

```python
df.head()
```

```python
df['vue_panoramique'] = df['vue_panoramique'] | df['vue_sur_mer']
```

```python
columns_to_drop = ['Amenities', 'Description', 'vue_sur_mer','Unnamed: 0', 'Title', 'URL','Description', 'Type', 'Condition',
       'Amenities', 'Published','Pieces']

df=df.drop(columns=columns_to_drop, axis=1)
```

```python
df.head()
```

```python
df = df.apply(lambda x: x.astype(int) if x.dtype == 'bool' else x)
```
